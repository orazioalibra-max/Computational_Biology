```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#np.random.seed(42)
r=3
nsim=1000
p_T=0
t_end=[]

for sim in range(nsim):
    t=0
    N=10
    t_plot=[0]
    N_plot=[N]

    while N>0:
        p_T = r*N # propensity to die

        x=np.random.uniform(0,1)
        ts=-np.log(x)/p_T
        t+=ts

        N-=1 # the only reaction in this setting (death)

        t_plot.append(t)
        N_plot.append(N)

    t_end.append(t)
    plt.figure(nsim)
    plt.plot(t_plot, N_plot)
    plt.xlabel('Time')
    plt.ylabel('Population')
    plt.title('Population trend')

mean_t=round(np.mean(t_end),4)
plt.figure(1)
plt.hist(t_end, bins=30, color='skyblue', edgecolor='black', alpha=0.7, density=True)
plt.xlabel('Extinction time')
plt.ylabel('Distribution')
plt.title('Distribution of extintion time, mean value = ' + str(mean_t))
plt.show()
```

Since death is a stochastic event, we cannot predict exactly when the next individual will die, but we know the probability of this event. At any moment, if there are N individuals and each has a death rate of k, the rate at which anyone in the population dies is N×k. As the population decreases, this rate slows down. As a result,the time between deaths follows an exponential distribution determined by the total rate. This means deaths happen quickly when N is high (start of simulation) and very slowly when N is low (end of simulation). This can be seen in the histogram up here. It shows the distribution of the extinction time and we can observe that while most populations go extinct around 1 day (the mean value of extintion time is around ~0.97 days), there is a long "tail" where some populations survive significantly longer because the final survivor (N=1) happened to live a long time.

We can conclude that stochasticity (randomness) plays a huge role in small populations, as seen also in the Moran process. Even with a high death rate, it is impossible to predict exactly when the population will vanish, but we can only predict the probability.
