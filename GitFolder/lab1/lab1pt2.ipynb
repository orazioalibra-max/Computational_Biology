```python
import numpy as np
import matplotlib.pyplot as plt

#np.random.seed(42)
t=np.arange(0,20000) #20 generations
pop_size=1000

a_size=int(pop_size/2)
b_size=int(pop_size/2)
a=np.zeros(a_size) #create population A
b=np.ones(b_size) #create population B
x=np.append(a,b)

# prob_a=a_size/(a_size+b_size)
# prob_b=b_size/(a_size+b_size)
lifetime=[]
birthtime=np.zeros(pop_size)
for i in t:
    dead_idx=np.random.randint(0, len(x))
    # if x[dead_idx] == 0:
    #     prob_a=prob_a-1/((a_size+b_size))
    # else:
    #     prob_b=prob_b-(1/(a_size+b_size))
    age=i-birthtime[dead_idx]
    lifetime.append(age)

    born_idx=np.random.randint(0, len(x))
    # if x[born_idx] == 0:
    #     prob_a=prob_a+1/((a_size+b_size))
    # else:
    #     prob_b=prob_b+(1/(a_size+b_size))
    x[dead_idx]=x[born_idx]
    birthtime[dead_idx]=i
plt.figure()
plt.hist(lifetime, bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Lifetime with pop_size='+str(pop_size))
plt.show()
```

Here I decided to plot timesteps to have a clearer histogram. As we can see, the lifetime follows an exponential trend with a <1 exponent. My guess is that it may be distributed as an exp~(1/pop_size) since the probability of an individual to die is independent from the age. To prove that, I'll plot such a distribution and compare the graphs.

```python
t_lin= np.linspace(0, max(lifetime), 100)
exp_dist = (1/pop_size) * np.exp(-t_lin/pop_size)
plt.figure()
plt.plot(t_lin, exp_dist, 'b-', lw=2)
plt.hist(lifetime, bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.show()
```
